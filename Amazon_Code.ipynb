```python
import matplotlib.pyplot as plt

# Calculate the value counts and percentages
region = df['Region'].value_counts()
percentages = region / region.sum() * 100

# Create the pie chart
colors = plt.cm.Paired(range(len(region)))  # Generate a list of colors

plt.figure(figsize=(8, 8))  # Set the figure size (optional)

# Create the pie chart without percentages displayed on the chart
wedges, texts = plt.pie(
    region,

    startangle=140,
    colors=colors
)

# Create a legend for category names separately
plt.legend(region.index, title="Units Sold", loc="upper left", bbox_to_anchor=(1, 1))

# Create a separate legend for percentage distribution
legend_labels = [f"{label}: {percent:.1f}%" for label, percent in zip(region.index, percentages)]
plt.legend(legend_labels, title="Percentage", loc="lower left", bbox_to_anchor=(1, 0))

plt.title('Regional Sales')

# Display the chart
plt.show()
```

```python
#CATEGORY
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each category
category_counts = df['Item Type'].value_counts()

# Create the bar graph
plt.figure(figsize=(12, 6))  # Set the figure size (optional)
category_counts.plot(kind='bar')
plt.title('Category Distribution')
plt.xlabel('Item Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
#QUANTITY
```

```python
import matplotlib.pyplot as plt

category_qty = df.groupby('Item Type')['Units Sold'].sum()
category_qty.plot(kind='bar', figsize=(8, 6))
plt.title('Total Units by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
#AMOUNT
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate statistics for the 'Amount' column
amount_stats = df['Unit Cost'].describe()

# Exclude the 'count' statistic
amount_stats = amount_stats.drop('count')

# Add median to the statistics table
amount_stats['Median'] = df['Unit Cost'].median()

# Create a custom statistics DataFrame
statistics_df = pd.DataFrame({
    'Statistic': amount_stats.index,
    'Value': amount_stats.values
})

# Plot the statistics
plt.figure(figsize=(10, 6))
plt.bar(statistics_df['Statistic'], statistics_df['Value'], color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Stats for Amount Column')
plt.xlabel('Statistic')
plt.ylabel('Value')

# Add text labels for each bar
for i, row in statistics_df.iterrows():
    plt.text(i, row['Value'], f'{row["Value"]:.2f}', ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df[['Units Sold', 'Unit Cost']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame for records with 'ship-city' as 'Bangalore'
bangalore_data = df[df['Region'] == 'Europe']

# Count the frequency of each category in Bangalore
category_counts = bangalore_data['Item Type'].value_counts()

# Set the figure size (optional)
plt.figure(figsize=(12, 6))

# Create a bar chart for category frequency in Bangalore
category_counts.plot(kind='bar', color='purple')
plt.title('Category Frequency in Europe')
plt.xlabel('Item Type')
plt.ylabel('Frequency')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=90)

# Show the bar chart
plt.show()
```
